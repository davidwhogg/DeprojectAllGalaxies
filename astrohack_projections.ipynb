```python
import astrohack_projections
reload(astrohack_projections)
import numpy as np

matplotlib.rcParams['figure.figsize'] = 20, 20

# build a kind of galaxy and project it onto x,y,z
gal_model_B = astrohack_projections.galaxy_model_3d()

scale = 4.
gal_model_B.add_gaussian(10.0, numpy.array((0.0, 0.0, 0.0)), numpy.array((1., 0, 0, 0, 2., 0, 0, 0, 3.)).reshape((3, 3)) * scale)
gal_model_B.add_gaussian(1.0, numpy.array((0.0, 0.0, 0.0)), numpy.eye(3) * 0.25 * scale)
gal_model_B.add_gaussian(1.0, numpy.array((0.0, 2.0, 2.0)), numpy.eye(3) * 0.25 * scale)
print type(gal_model_A.get_total_mass())

gal_model_B = gal_model_B * (gal_model_A.get_total_mass() / gal_model_B.get_total_mass())

# now do the projection 
xs = numpy.arange(-20, 20, 0.2)
ys = numpy.arange(-20, 20, 0.2)
Y, X = numpy.meshgrid(ys, xs)

# plot the same galaxy but with different rotation values
x_angles = numpy.array((0.0, 30.0, 60.0, 90.0))
y_angles = numpy.array((0.0, 30.0, 60.0, 90.0))

kwargs = {"interpolation": "nearest",
          "cmap": "afmhot",
          "origin": "lower"}
psf = astrohack_projections.mixture_of_gaussians(2)
psf.add_gaussian(1., np.array([0., 0.]), np.eye(2)*1.)

numpy.random.seed(42)
images = []
metadata = []
for i, x_angle in enumerate(x_angles):
    for j, y_angle in enumerate(y_angles):
        
        xi_hat, eta_hat = astrohack_projections.choose_random_projection()
        intensity = 1.
        scale = 0.25 * np.exp(np.random.uniform())
        xshift = np.random.uniform(13., 16.)
        yshift = np.random.uniform(18., 21.)
        
        image = astrohack_projections.image_model()
        image.set_shape((30, 40))
        if j % 2 == 0:
            gal = gal_model_A
        else:
            gal = gal_model_B
        image.set_psf(psf)
        image.add_galaxy(gal, xi_hat, eta_hat, intensity, scale, xshift, yshift)
        image.set_ivar(np.ones(image.get_shape()))
        image.add_random_noise() 
        
        a, mid = np.percentile(image.get_data(), [25., 50.])
        ran = mid - a
        vmin = mid - 5. * ran
        vmax = mid + 20. * ran
        
        plt.subplot(len(x_angles), len(y_angles), 4*i + j + 1)
        plt.imshow(image.get_data(), vmin=vmin, vmax=vmax, **kwargs)
        images.append(image)
        metadata.append((scale, xi_hat, eta_hat, xshift, yshift))
        plt.colorbar()
        
```

```python
matplotlib.rcParams['figure.figsize'] = 8, 4

ii=14
print metadata[ii]

def showme(img):
    a, mid = np.percentile(img.get_data(), [25., 50.])
    ran = mid - a
    vmin = mid - 5. * ran
    vmax = mid + 20. * ran
    plt.subplot(1, 2, 1)
    plt.title("data")
    plt.imshow(img.get_data(), vmin=vmin, vmax=vmax, **kwargs)
    plt.subplot(1, 2, 2)
    plt.title("synthetic")
    plt.imshow(img.get_synthetic(), vmin=vmin, vmax=vmax, **kwargs)

showme(images[ii])
```
